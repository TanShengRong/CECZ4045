```python
import matplotlib.pyplot as plt
```

```python
import os

valid_ar = [[None]]*6

for i in range(1,7):
    path = "./cnn_layer_{}.txt".format(str(i))
    if os.path.exists(path):
        with open(path,"r") as f:
            lines = f.read().splitlines()
            lines = [float(line) for line in lines]
            valid_ar[i-1] = lines 

print(valid_ar[0])
```

```python
plt.plot(valid_ar[0])
plt.show()
```

```python
plt.rcParams['figure.dpi'] = 80
plt.style.use('seaborn-pastel')

plt.plot(valid_ar[0], label="1 Layer CNN")
plt.plot(valid_ar[1], label="2 Layer CNN")
plt.plot(valid_ar[2], label="3 Layer CNN")
plt.plot(valid_ar[3], label="4 Layer CNN")
plt.plot(valid_ar[4], label="5 Layer CNN")
plt.plot(valid_ar[5], label="6 Layer CNN")

# plt.tick_params(labelleft=False, left=False)
plt.ylim(0.86,0.94)

plt.title("F-1 Score on Validation Set vs Time Intervals")
plt.xlabel("Time Intervals")
plt.ylabel("F-1")
plt.legend()
plt.show()
```
